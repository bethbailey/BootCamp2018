```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Make list of xs to plot
x = np.linspace(0, 50, 1000)

# Find constraint line ys on those xs
con1 = (2 * x + 4) / 3
con2 = (x - 1) / 6
con3 = 6 - x
```

```python
plt.figure(figsize=(10, 10))

plt.plot(x, con1)
plt.plot(x, con2)
plt.plot(x, con3)
plt.axvline(x=0)
plt.plot(x, np.zeros(1000))
plt.xlim(-1, 8)
plt.ylim(-1, 8)
# Put a label on and fill in the part designating the feasible set
plt.text(1, 3.5, "Feasible", fontsize = 15)
plt.fill_between(x, con3, con1, where = con3 >= con1)

# Plot the objective function over the feasible set
obj1 = (5 * x - 10) / 4
obj2 = (5 * x - 20) / 4
obj3 = (5 * x) / 4
obj4 = (5 * x + 10) / 4
plt.plot(x, obj1, 'r--')
plt.plot(x, obj2, 'r--')
plt.plot(x, obj3, 'r--')
plt.plot(x, obj4, 'r--')
# Plot the optimizer
plt.plot(2.8, 3.2, 'o')

plt.xlabel('x values')
plt.ylabel('y values')

plt.show()
```

The optimum is at x = 2.8, y = 3.2.

## 8.15

Weak Duality Theorem: If $x \in \mathbb{R}^n$ is a primal feasible point and $y \in \mathbb{R}^m$ is any feasible point of the dual, then $c^Tx \leq b^Ty$. 

Let $x \in \mathbb{R}^n$ be a primal feasible point and $y \in \mathbb{R}^m$ be any feasible pont of the dual. Then $Ax \preceq b$ and $A^Ty \succeq b$, and therefore we know $(Ax)^T y \leq b^T y$ and $x^TA^Ty \geq x^Tc$. Thus, we have the below:

$$c^T x = x^T c$$
$$\leq x^TA^T y$$
$$= (Ax)^T y$$
$$\leq b^T y$$

## 8.17

Let the primal problem be as follows:

$$max c^Tx$$
$$s.t. Ax \preceq b$$
$$x \succeq 0$$

Then we have the dual as follows:

$$max b^Ty$$
$$s.t. A^Ty \succeq c$$
$$y \succeq 0$$

Thus, the dual of the dual (by definition) is:

$$max c^Tz$$
$$s.t. (A^T)^Tz \succeq b$$
$$z \succeq 0$$

And we can see that this is the primal problem, since $(A^T)^T = A$.

$$max c^Tz$$
$$s.t. Az \succeq b$$
$$z \succeq 0$$
