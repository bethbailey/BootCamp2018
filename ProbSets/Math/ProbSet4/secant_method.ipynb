```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def secant_min(x0, x1, err, fp):
    
    maxIter = 1000
    iters = 0
    xk = x1
    xkmin1 = x0
    diff = err + 1
    x_conv = [x0, x1]
    converged = True
    
    while iters < maxIter and diff > err:
        x = xk - fp(xk) * (xk - xkmin1) / ( fp(xk) - fp(xkmin1) )
        xkmin1 = xk
        xk = x
        x_conv.append(xk)
        diff = abs(xk - xkmin1) / abs(xkmin1)
        iters += 1
    
    if iters == maxIter:
        converged = False
        
    return converged, x_conv, diff, iters
```

```python
fp = lambda x: 4 * x**3 - 14 * 3 * x**2 + 120 * x - 70
```

```python
x0 = 0
x1 = 0.1
err = 10e-8
```

```python
conv, x_vec, diffs, iters = secant_min(x0, x1, err, fp)
```

```python
if conv == True:
    print("The solution converged at error rate", diffs, "after ", iters, "iterations.")
else:
    print("The solution did not converge.")
```

```python
# Plot the function
x_vec = np.array(x_vec)

f = lambda x: x**4 - 14 * x**3 + 60 * x**2 - 70 * x

fig, ax = plt.subplots(figsize=(8, 8))
x_grid = np.linspace(0, 2)
ax.plot(x_grid, f(x_grid))
ax.scatter(x_vec, f(x_vec))
plt.title("Secant Method Minimizer")
plt.show()
```
